```python
# Metrics Tool
def analyze_metrics(file_path):
    df = pd.read_csv(file_path)

    trends = {}
    for col in df.columns[1:]:
        trends[col] = df[col].iloc[-1] - df[col].iloc[0]

    return trends


# Sentiment Tool
def analyze_sentiment(feedback_list):
    negative_keywords = ["crash", "fail", "slow", "bug", "error"]

    negative = 0
    for fb in feedback_list:
        if any(word in fb.lower() for word in negative_keywords):
            negative += 1

    score = negative / len(feedback_list)

    return {
        "negative_ratio": score,
        "summary": "High negative sentiment" if score > 0.5 else "Mixed sentiment"
    }

print("Tools ready ")
```

```python
# Data Analyst Agent
class DataAnalystAgent:
    def run(self, metrics_path):
        trends = analyze_metrics(metrics_path)
        issues = [k for k, v in trends.items() if v < 0]

        return {
            "trends": trends,
            "issues": issues,
            "summary": "Metrics degrading" if issues else "Metrics stable"
        }


# Marketing Agent
class MarketingAgent:
    def run(self, feedback_path):
        import json
        feedback = json.load(open(feedback_path))
        return analyze_sentiment(feedback)


# PM Agent
class PMAgent:
    def run(self, metrics_insights, sentiment_insights):
        if len(metrics_insights["issues"]) > 2 and sentiment_insights["negative_ratio"] > 0.5:
            decision = "ROLLBACK"
        elif sentiment_insights["negative_ratio"] > 0.4:
            decision = "PAUSE"
        else:
            decision = "PROCEED"

        return {
            "decision": decision,
            "rationale": "Based on degrading metrics and user sentiment"
        }


# Critic Agent
class CriticAgent:
    def run(self, decision):
        risks = []

        if decision == "PROCEED":
            risks.append("Ignoring negative feedback may harm users")
        elif decision == "PAUSE":
            risks.append("Delay may impact business goals")
        elif decision == "ROLLBACK":
            risks.append("Feature investment wasted temporarily")

        return risks

print("Agents ready ")
```

```python
import json

# Initialize agents
data_agent = DataAnalystAgent()
marketing_agent = MarketingAgent()
pm_agent = PMAgent()
critic_agent = CriticAgent()

print("War Room Simulation Started...\n")

# Step 1
metrics_insights = data_agent.run("data/metrics.csv")
print(" Metrics Insights:", metrics_insights)

# Step 2
sentiment_insights = marketing_agent.run("data/feedback.json")
print("💬 Sentiment Insights:", sentiment_insights)

# Step 3
decision_data = pm_agent.run(metrics_insights, sentiment_insights)
print(" PM Decision:", decision_data)

# Step 4
risks = critic_agent.run(decision_data["decision"])
print("Risks:", risks)

# Final Output
output = {
    "decision": decision_data["decision"],
    "rationale": decision_data["rationale"],
    "risks": risks,
    "confidence": 0.75
}

with open("output/result.json", "w") as f:
    json.dump(output, f, indent=4)

print("\n Output saved in output/result.json")
```
